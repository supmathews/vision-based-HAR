```python
# function to display random frames of videos from the dataset
def display_random_frames(dir):
  plt.figure(figsize=(30, 30))

  # get the names of classes
  classes_list = os.listdir(dir)

  # generate a list of random frames
  # range of the list
  random_range = random.sample(range(len(classes_list)), 2)

  # iterate through the random_range values
  for c, i in enumerate(random_range, 1):

    # get class name of the index
    selected_class_name = classes_list[i]

    # get the list of all videos belonging to that class
    video_name_list = os.listdir(f'{dir}/{selected_class_name}')

    # select a random video from the list
    selected_video = random.choice(video_name_list)

    # initialize opencv to read in the video
    video_reader = cv2.VideoCapture(f'{dir}/{selected_class_name}/{selected_video}')

    # get the first frame of the video
    _, bgr_frame = video_reader.read()

    # release the VideoCapture object
    video_reader.release()

    # convert bgr frame to rgb frame
    rgb_frame = cv2.cvtColor(bgr_frame, cv2.COLOR_BGR2RGB)

    # add class name to frame 
    cv2.putText(rgb_frame, selected_class_name, (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # display the frame
    plt.subplot(5, 4, c);
    plt.imshow(rgb_frame);
    plt.axis('off')
```

```python
# function to extract the features from the video
# source : https://bleedai.com/human-activity-recognition-using-tensorflow-cnn-lstm/
def frames_extraction(video_path):

  # list to store the frames
  frames_list = []

  # read the video file
  video_reader = cv2.VideoCapture(video_path)

  # total number of frames in video
  video_frame_count = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))

  # calculate the interval between each frame that is in list
  frame_interval = max(int(video_frame_count/SEQUENCE_LENGTH), 1)

  # iterate through the frames collected
  for counter in range(SEQUENCE_LENGTH):

    # set the position of the frame and read the frame
    video_reader.set(cv2.CAP_PROP_POS_FRAMES, counter*frame_interval)
    success, frame = video_reader.read()

    # if not success break 
    if not success:
      break
    
    # resize the frame and normalize the frame so that pixel value is between 0 and 1
    resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))
    normalized_frame = resized_frame / 255

    # append the normalized frames to frames list
    frames_list.append(normalized_frame)
  
  video_reader.release()

  return frames_list
```

```python
# function to extract features, labels and video paths from the dataset
# source : https://bleedai.com/human-activity-recognition-using-tensorflow-cnn-lstm/
def create_dataset():

  # list to store the features, labels and video paths of the videos
  features = []
  labels = []
  video_file_paths = []

  for id, name in enumerate(CLASSES_LIST):
    print('Extracting data of classs : ', name)

    # get the list of video files in the directory
    files_list = os.listdir(os.path.join(DATASET_DIR, name))

    # iterate through files_list
    for file_name in files_list:
      # get the file path of the video
      video_file_path = os.path.join(DATASET_DIR, name, file_name)

      # call the frames_extraction method to extract the frames
      frames = frames_extraction(video_file_path)

      # if video is shorter than SEQUENCE_LENGTH ignore them
      if len(frames) == SEQUENCE_LENGTH:
        features.append(frames)
        labels.append(id)
        video_file_paths.append(video_file_path)
      
  # convert list to numpy arrays
  features = np.asarray(features)
  labels = np.array(labels)  

  return features, labels, video_file_paths
```

```python
from keras.backend import shape
# model proposed in "A New Approach for Abnormal Human Activities Recognition Based on ConvLSTM Architecture"
# source : https://www.mdpi.com/1424-8220/22/8/2946

def create_cnn_convlstm_model():
  
  # define a sequential model
  model = Sequential(name='cnn_convlstm_model')

  ############ model architecture ##############

  ############ input layer #####################
  model.add(Input(shape=(SEQUENCE_LENGTH, IMAGE_HEIGHT, IMAGE_WIDTH, 3)))
  # define a time distributed conv2d layer
  model.add(TimeDistributed(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu')))

  ############ hidden layers ###################
  # define convlstm layer
  model.add(ConvLSTM2D(filters=64, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True, data_format = "channels_last"))

  # define batch normalization layer
  model.add(BatchNormalization())

  # define standard conv2d layer
  model.add(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu'))

  # define dropout of 50%
  model.add(Dropout(0.5))

  # define flatten layer
  model.add(Flatten())

  # define dense layer
  model.add(Dense(units=256, activation='relu'))

  # define dropout of 50% 
  model.add(Dropout(0.5))

  ########## output layer ####################
  # define dense layer
  model.add(Dense(len(CLASSES_LIST), activation='sigmoid')) 

  ############################################
  model.build()
  # display model summary
  model.summary()

  return model
```

```python
from keras.backend import shape
# adding extra maxpool layers and drop out layers to original model

def create_cnn_convlstm_model_mod_2():
  
  # define a sequential model
  model = Sequential(name='cnn_convlstm_model')

  ############ model architecture ##############
  model.add(TimeDistributed(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu'), input_shape=(SEQUENCE_LENGTH, IMAGE_HEIGHT, IMAGE_WIDTH, 3)))

  model.add(AveragePooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last'))

  # define convlstm layer
  model.add(ConvLSTM2D(filters=64, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True, data_format = "channels_last"))

  model.add(ConvLSTM2D(filters=128, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True, data_format = "channels_last"))

  model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last'))
  model.add(AveragePooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last'))
  model.add(TimeDistributed(Dropout(0.2)))

  # define batch normalization layer
  model.add(BatchNormalization())

  model.add(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu'))

  model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last'))
  model.add(TimeDistributed(Dropout(0.2)))

  model.add(Flatten())

  model.add(Dense(units=256, activation='relu'))

  # define dropout of 50% 
  model.add(Dropout(0.5))

  ########## output layer ####################
  # define dense layer
  model.add(Dense(len(CLASSES_LIST), activation='signmoid')) 

  ############################################
  #model.build()
  # display model summary
  model.summary()
  return model
```

```python
# function to predict on selected video, apply label to frames and generate the video
# source : https://bleedai.com/human-activity-recognition-using-tensorflow-cnn-lstm/
def predict_on_video(model, video_file_path, output_file_path, SEQUENCE_LENGTH):
 
    # initialize the VideoCapture object to read from the video file.
    video_reader = cv2.VideoCapture(video_file_path)
 
    # get the width and height of the video.
    original_video_width = int(video_reader.get(cv2.CAP_PROP_FRAME_WIDTH))
    original_video_height = int(video_reader.get(cv2.CAP_PROP_FRAME_HEIGHT))
 
    # initialize the VideoWriter Object to store the output video in the disk.
    video_writer = cv2.VideoWriter(output_file_path, cv2.VideoWriter_fourcc('M', 'P', '4', 'V'), 
                                   video_reader.get(cv2.CAP_PROP_FPS), (original_video_width, original_video_height))
 
    # declare a queue to store video frames.
    frames_queue = deque(maxlen = SEQUENCE_LENGTH)
 
    # initialize a variable to store the predicted action being performed in the video.
    predicted_class_name = ''
 
    # iterate until the video is accessed successfully.
    while video_reader.isOpened():
 
        # read the frame.
        ok, frame = video_reader.read() 
        
        # check if frame is not read properly then break the loop.
        if not ok:
            break
 
        # resize the Frame to fixed Dimensions.
        resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))
        
        # normalize the resized frame by dividing it with 255 so that each pixel value then lies between 0 and 1.
        normalized_frame = resized_frame / 255
 
        # appending the pre-processed frame into the frames list.
        frames_queue.append(normalized_frame)
 
        # check if the number of frames in the queue are equal to the fixed sequence length.
        if len(frames_queue) == SEQUENCE_LENGTH:
 
            # pass the normalized frames to the model and get the predicted probabilities.
            predicted_labels_probabilities = model.predict(np.expand_dims(frames_queue, axis = 0))[0]
 
            # get the index of class with highest probability.
            predicted_label = np.argmax(predicted_labels_probabilities)
 
            # get the class name using the retrieved index.
            predicted_class_name = CLASSES_LIST[predicted_label]
 
        # write predicted class name on top of the frame.
        cv2.putText(frame, predicted_class_name, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
 
        # write The frame into the disk using the VideoWriter Object.
        video_writer.write(frame)
        
    # release the VideoCapture and VideoWriter objects.
    video_reader.release()
    video_writer.release()
```

```python
# function to plot accuray/loss graphs
def plot_graph(history, n, epochs):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(epochs)
    
    plt.figure(figsize=(9, 9))
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='best')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy')
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='best')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.savefig(str(n) + '.png')
    plt.show()
```

```python
# function to attach the pre-trained model to custom model

def attach_base_model(base_model):
  ip = Input(shape=(SEQUENCE_LENGTH, IMAGE_HEIGHT, IMAGE_WIDTH, 3))

  x = TimeDistributed(base_model)(ip)

  x = TimeDistributed(Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu'))(x)

  x = AveragePooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last')(x)

  x = ConvLSTM2D(filters=64, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True, data_format = "channels_last", padding='same')(x)

  x = ConvLSTM2D(filters=64, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True, data_format = "channels_last", padding='same')(x)

  X = MaxPooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last')(x)

  x = AveragePooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last')(x)

  x = TimeDistributed(Dropout(0.2))(x)

  x = BatchNormalization()(x)

  x = Conv2D(filters=16, kernel_size=(3, 3), padding='same', activation='relu')(x)

  x = MaxPooling3D(pool_size=(1, 2, 2), padding='same', data_format='channels_last')(x)

  x = TimeDistributed(Dropout(0.2))(x)

  x = Flatten()(x)

  x = Dense(units=256, activation='relu')(x)

  x = Dropout(0.5)(x)

  x = Dense(len(CLASSES_LIST), activation='sigmoid')(x)

  model = Model(ip, x)

  return model
```

```python
# get the final accuracy, precision, recall, f1 score, roc-auc-curve score
def evaluate_model(y_true, y_pred, y_score):
  acc = accuracy_score(y_true, y_pred)
  precision = precision_score(y_true, y_pred, zero_division=1)
  recall = recall_score(y_true, y_pred)
  f1 = f1_score(y_true, y_pred)
  auc = roc_auc_score(y_true, y_score)
  print('>Accuracy:', acc)
  print('>Precision: ', precision)
  print('>Recall: ', recall)
  print('>f1-Score: ', f1)
  print('>roc-auc-curve-score: ', auc)
```
